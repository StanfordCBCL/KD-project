Implementing and visualizing secondary flow calculations from Banko et al. 2015. 

The formulation looks like this: 

$$I_1 = \Big( \frac{ \int \int_A (\mathbf{u} \cdot \hat{\mathbf{n}})^2 dA}{U_T^2 A} \Big)^{\frac{1}{2}}$$

$$I_2 = \Big( \frac{ \int \int_A (\mathbf{u} - (\mathbf{u} \cdot \hat{\mathbf{n}})\hat{\mathbf{n}} )^2 dA}{U_T^2 A} \Big)^{\frac{1}{2}}$$

Import packages, define utility functions first

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os 
import pickle
%matplotlib inline
````

````python
def read_from_file(name): 
    path = os.getcwd() + '/AneurysmGeneration/pickles/' + name + '.pkl'
    file = open(path, 'rb')
    new_obj = pickle.load(file)
    file.close()
    return new_obj
````

Load results that were previously computed through ```second_flows.py```

````python
results = read_from_file('second_flows_dict_2')
````

Select the aneurysm to be visualized. 

Note that for each aneurysm, secondary flows were computed at 6 regularly-spaced slices throughout the length of the aneurysm, in planes sliced orthogonal to the vessel centerline. The results are stored in a dictionary such that the position/size of the aneurysm is the key, mapping to all secondary flows computed for that aneurysm. 

The value contains the following: 
* ```all_second_flows```: integrated numerator of the $I_2$ formulation, computed at each slice.  
* ```areas```: $A$, the aneurysm cross sectional areas for each plane through which the secondary flow was computed. From the denominator of the $I_1$ and $I_2$ formulations. 
* ```centers```: the vessel centerline point for each plane through which the secondary flow was computed. 
* ```normals```: the normal vector for each plane through which the secondary flow was computed. Used as $\hat{\mathbf{n}}$ in $I_1$ and $I_2$ formulations.

Note that there are secondary flows computed at 6 slices for each aneurysm (axis 0) and $\mathbf{u}$ varies with time (cardiac cycle divided into 20 steps, axis 1). 

Because the pre-computed results only contained the numerator, we are going to need to compute the denominator and divide by this result.

````python
all_second_flows = np.array(all_second_flows)
all_second_flows.shape
````

````python
np.divide(np.square(velocity_peaks), areas).shape
# np.divide(all_second_flows, )
# all_second_flows = np.divide(all_second_flows, np.array(areas).reshape(len(areas), 1))
normalized_second_flows = np.divide(all_second_flows.T, np.divide(np.square(velocity_peaks), areas))
````

````python
normalized_second_flows = np.transpose(normalized_second_flows)
````

Let's plot 

````python
positions = 6
interval = 1./(positions+1)
x_pos = [i*interval for i in range(1, positions+1)]

step_lower = 3000
step_upper = 4000
tstep = 50
steps = np.arange(step_lower, step_upper + tstep, tstep)

step_labels = ['velocity_0' + str(step) for step in steps]

fig = plt.figure(figsize=(10,10))
for i, (second_flow, label) in enumerate(zip(np.transpose(normalized_second_flows), step_labels)): 
    if i % 2 == 0: 
        continue
    plt.plot(x_pos, second_flow, label = label)

plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.xlabel('Position along aneurysm')
plt.ylabel(r'I_2')
plt.title(which_aneurysm + ', with area normalization')
````

````python
fig, axs = plt.subplots(3, 2, figsize=(10,10))
axs[0,0].plot(steps, normalized_second_flows[0,:])
axs[1,0].plot(steps, normalized_second_flows[1,:])
axs[2,0].plot(steps, normalized_second_flows[2,:])
axs[0,1].plot(steps, normalized_second_flows[3,:])
axs[1,1].plot(steps, normalized_second_flows[4,:])
axs[2,1].plot(steps, normalized_second_flows[5,:])

axs[0,0].title.set_text('Position 1')
axs[1,0].title.set_text('Position 2')
axs[2,0].title.set_text('Position 3')
axs[0,1].title.set_text('Position 4')
axs[1,1].title.set_text('Position 5')
axs[2,1].title.set_text('Position 6')

for ax_list in axs:
    for ax in ax_list:
        ax.set_ylim([0, 8])
````
